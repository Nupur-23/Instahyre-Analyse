```python
def temp(val):
    if val == 1:
        return  '0-10'
    elif val == 10:
        return '10-50'
    elif val == 50:
        return '50-200'
    elif val == 200:
        return '200-500'  
    else:
        return '>1000'
```

```python
def jobs(val):
    val = val.split(' ')
    if 'Software' in val :
        return 'Software'
    elif 'Data' in val :
        return 'Data'
    elif 'Business' in val:
        return 'Business'
    elif 'Frontend' in val:
        return 'Web'
    elif 'Backend' in val:
        return 'Web'
    elif 'Stack' in val:
        return 'Web'
    elif 'DevOps' in val:
        return 'DevOps'
    elif 'Sales' in val:
        return 'Sales'
    elif 'Marketing' in val:
        return 'Marketing'
    elif 'Cloud' in val:
        return 'cloud'
    else:
        return 'other'
```

```python
df['industry'] = df['job_title'].map(jobs)
```

```python
import requests
from bs4 import BeautifulSoup

def extract_company_names(url):
    # Send a GET request to the website
    response = requests.get(url)
    
    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the relevant HTML elements containing the company names
    company_elements = soup.select('.position-company')
    
    # Extract the company names from the HTML elements
    company_names = [element.text.strip() for element in company_elements]
    
    return company_names

# Example usage
url = 'https://www.instahyre.com/search-jobs/'
companies = extract_company_names(url)
print(companies)
```
